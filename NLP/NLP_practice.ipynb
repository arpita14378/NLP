```python
import nltk
nltk.download('gutenberg')
nltk.corpus.gutenberg.fileids()
```

```python
book1 = nltk.corpus.gutenberg.words( 'carroll-alice.txt')
first500 = book1[0:500]
first500
```

```python
#step - 1 ->to remove all punctuations and non words
norm_1 = [words for words in first500 if words.isalpha()]
norm_1
```

```python
#step-2 ->remove stop words
from nltk.corpus import stopwords
nltk.download("stopwords")
norm_2 = [word for word in norm_1 if word not in stopwords.words('english')]
norm_2
```

```python
#step-3 ->to lowercase
norm_3 = [word.lower() for word in norm_2]
norm_3
```

```python
#step-4 ->use Stemmer
stemmer = nltk.PorterStemmer()

norm_4 = [stemmer.stem(word) for word in norm_3]

for i in range(0,len(norm_3)):
    if norm_3[i]!=norm_4[i]:
        print(norm_3[i],norm_4[i])
```

```python
#step-5 ->use Lemmatization

lemmatizer = nltk.WordNetLemmatizer()
nltk.download('wordnet')
norm_5 = [lemmatizer.lemmatize(words) for words in norm_4] 

for i in range(0,len(norm_4)):
    if norm_4[i]!=norm_5[i]:
        print(norm_4[i],norm_5[i])
```
